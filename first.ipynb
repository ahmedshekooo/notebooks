```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("./vgsales.csv")
df.head()
```

```python
df.hist(figsize=(10,10),color='#151515',bins=20)
```

### first question:

#### what is the most popular publisher?


```python
most_popular_publisher = df.groupby('Publisher')['Global_Sales'].sum().idxmax()

publisher_df = df[df['Publisher'] == most_popular_publisher]

publisher_df['Genre'].value_counts().plot(kind='bar', figsize=(10, 6), color='#151515')
plt.title(f'Number of Games by Genre for {most_popular_publisher}')
plt.xlabel('Genre')
plt.ylabel('Number of Games')
plt.show()
```
