```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("./vgsales.csv")
df.head()
```

```python
df.hist(figsize=(10,10),color='#151515',bins=20)
```

```python
plt.scatter(df['Year'],df['Global_Sales'],color='#ffda22')
```

<h3 style="color:#ffda22"> first question </h3>

#### what is the most popular publisher?


```python
most_popular_publisher = df.groupby('Publisher')['Global_Sales'].sum().idxmax()

publisher_df = df[df['Publisher'] == most_popular_publisher]

publisher_df['Genre'].value_counts().plot(kind='bar', figsize=(10, 6), color='#151515')
plt.title(f'Number of Games by Genre for {most_popular_publisher}')
plt.xlabel('Genre')
plt.ylabel('Number of Games')
plt.show()
```

<h3 style="color:#ffda22"> second question </h3>

#### what is the most popular publisher in USA ?

```python
most_popular_publisher = df.groupby('Publisher')['NA_Sales'].sum().idxmax()

publisher_df = df[df['Publisher'] == most_popular_publisher]

publisher_df['Genre'].value_counts().plot(kind='bar', figsize=(10, 6), color='#151515')
plt.title(f'Number of Games by Genre for {most_popular_publisher}')
plt.xlabel('Genre')
plt.ylabel('Number of Games')
plt.show()
```

<h3 style="color:#ffda22"> third question </h3>

#### what is the most popular type of game

```python
most_popular_genre = df.groupby('Genre')['NA_Sales'].sum().idxmax()

genre_df = df[df['Genre'] == most_popular_genre]
genre_publishers = genre_df['Publisher'].value_counts()
genre_publishers = genre_publishers.head(10)
genre_publishers.plot(kind='bar', figsize=(10, 6), color='#151515')
plt.title(f'Number of Publishers for {most_popular_genre} Games')
plt.xlabel('Publisher')
plt.ylabel('Number of Games')
plt.show()
```

<h3 style="color:#ffda22"> fourth question </h3>

#### which is more effecting on global sales publisher or genre

```python
nintendo_global_sales = publisher_df['Global_Sales'].sum()

action_global_sales = genre_df['Global_Sales'].sum()

print(f"Total Global Sales for Nintendo Publisher: {nintendo_global_sales}")
print(f"Total Global Sales for Action Genre: {action_global_sales}")
```

<h3 style="color:#ffda22">so that publisher is more effecting on global sales</h3>
